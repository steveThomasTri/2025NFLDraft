## Since this is confirmed, now I can get tables from 2000-2024

```python
combinetable = ""

for x in range (2000,2025):
    browser.visit(f"https://www.pro-football-reference.com/draft/{x}-combine.htm")
    table = browser.find_by_css(".table_container")[0]
    parsedtable = pd.read_html(table.html)[0]
    parsedtable["Year"] = x

    if x == 2000:
        combinetable = parsedtable
    else:
        combinetable = pd.concat([combinetable,parsedtable], ignore_index=True)

combinetable
```

## Data Wrangling

```python
combinetable = combinetable.drop(["College"], axis='columns')
```

```python
combinetable # with College column removed
```

```python
#only rows where a prospect completed in at least one combine event
deleteallnarows = combinetable[(combinetable["Bench"].isna()) & \
(combinetable["40yd"].isna()) & \
(combinetable["Vertical"].isna()) & \
(combinetable["Broad Jump"].isna()) & \
(combinetable["3Cone"].isna()) & \
(combinetable["Shuttle"].isna())].index

deleteallnarows
```

```python
combinetable.drop(deleteallnarows , inplace=True)
```

```python
combinetable # only rows that have at least one event
```

```python
drafted_df = combinetable.copy()
```

```python
drafted_df
```

```python
drafted_df = drafted_df.rename(columns={'Drafted (tm/rnd/yr)': 'isDrafted'})
drafted_df
```

```python
# Replace values in 'col1' where 'col1' is less than 3
condition2 = drafted_df['isDrafted'].isna()
drafted_df.loc[condition2, 'isDrafted'] = 0

# Replace values in 'col2' based on a condition in 'col1'
condition3 = drafted_df['isDrafted'] != 0
drafted_df.loc[condition3, 'isDrafted'] = 1
```

```python
drafted_df
```

```python
#only rows where a prospect completed in at least one combine event
deleteHT = drafted_df[drafted_df["Ht"] == "Ht"].index
deleteHT

drafted_df.drop(deleteHT , inplace=True)
```

```python
drafted_df
```

```python
## we looked at Excel file and we will need to drop all of those players with no height.  They are all from 2021.
#only rows where a prospect completed in at least one combine event
deleteHT = drafted_df[drafted_df["Ht"].isna()].index
deleteHT

drafted_df.drop(deleteHT , inplace=True)
```

```python
drafted_df
```

```python
count = 0

for index,row in drafted_df.iterrows():
    if "-" in row["Ht"]:
        count = count + 1

count
#This is now verified
```

```python
drafted_df.to_csv("drafted_2000_2024-prospectswithnoheightremoved.csv")
```

```python
for index,row in drafted_df.iterrows():
    if "-" in row["Ht"]:
        feet = int(row["Ht"].split("-")[0])
        inches = int(row["Ht"].split("-")[1])
        #print(feet * 12 + inches)
        drafted_df.loc[index,"Ht"] = feet * 12 + inches
```
