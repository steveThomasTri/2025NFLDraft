```python
import pandas as pd
from xgboost import XGBClassifier, plot_importance
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

df = pd.read_csv("cleaned_NFL_combine_2000_to_2024.csv")
df = df[(df["position"] != "NT") & (df["position"] != "K") & (df["position"] != "LS")]
df = df.drop(["year", "name", "college", "position"], axis=1)
df["drafted"] = df["drafted"].replace({"Y": 1, "N": 0})
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
df
```

```python
# Optional: handle missing values
# df = df.dropna()  # OR use imputation instead

Y = df["drafted"]
#df = df.apply(pd.to_numeric, errors='coerce')
X = df.drop(["drafted"], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=777)

model = XGBClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
# make predictions for test data
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
```

```python
# evaluate predictions
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
# fit model on training data
model = XGBClassifier(max_depth=3)
model.fit(X_train, y_train)

# make predictions for test data
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]

# evaluate predictions
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
