```python
from splinter import Browser
from sys import platform
import os
import csv

import pandas as pd
import numpy as np
```

```python
from selenium.webdriver.chrome.service import Service

# Optional: set path to ChromeDriver if not in PATH
service = Service(executable_path='chromedriver.exe')
browser = Browser('chrome', service=service)
```

```python
# https://www.pro-football-reference.com/draft/2025-combine.htm
browser.visit("https://www.pro-football-reference.com/draft/2025-combine.htm")
```

```python
players = browser.find_by_xpath('//*[@data-stat="player"]')
poss = browser.find_by_xpath('//*[@data-stat="pos"]')
schoolnames = browser.find_by_xpath('//*[@data-stat="school_name"]')
heights = browser.find_by_xpath('//*[@data-stat="height"]')
weights = browser.find_by_xpath('//*[@data-stat="weight"]')
fortyyds = browser.find_by_xpath('//*[@data-stat="forty_yd"]')
verticals = browser.find_by_xpath('//*[@data-stat="vertical"]')
benchrepss = browser.find_by_xpath('//*[@data-stat="bench_reps"]')
broadjumps = browser.find_by_xpath('//*[@data-stat="broad_jump"]')
cones = browser.find_by_xpath('//*[@data-stat="cone"]')
shuttles = browser.find_by_xpath('//*[@data-stat="shuttle"]')
```

```python
prospects = []
positions = []
school_names = []
hts = []
wts = []
forty_yds = []
vertical = []
bench_reps = []
broad_jumps = []
cone = []
shuttle = []
```

```python
for player in players:
    prospects.append(player.text)
for position in poss:
    positions.append(position.text)
for school in schoolnames:
    school_names.append(school.text)
for height in heights:
    hts.append(height.text)
for weight in weights:
    wts.append(weight.text)
for f in fortyyds:
    forty_yds.append(f.text)
for v in verticals:
    vertical.append(v.text)
for bench in benchrepss:
    bench_reps.append(bench.text)
for jump in broadjumps:
    broad_jumps.append(jump.text)
for c in cones:
    cone.append(c.text)
for s in shuttles:
    shuttle.append(s.text)
```

```python
len(shuttle)
```

```python
prospect_info = pd.DataFrame({"player":prospects,"college":school_names,"position":positions,"height":hts,'weight':wts,'40yd':forty_yds,'vertical':vertical,'bench reps':bench_reps,'broad jump':broad_jumps,'cone':cone,'shuttle':shuttle})
prospect_info
```

```python
prospect_info = prospect_info[prospect_info["player"] != "Player"]
```

```python
prospect_info.to_csv("prospectinfo.csv", index=False)
```
